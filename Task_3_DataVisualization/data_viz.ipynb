In this notebook, we transform HR data into visual insights using Python libraries.  
The aim is to create charts that reveal relationships and patterns around employee attrition.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("HR Analytics.csv")
df.head()
```

ðŸ”¹ 1. Attrition Count

```python
sns.countplot(x='Attrition', data=df, palette='Set2')
plt.title("Overall Attrition Count")
plt.show()
```

ðŸ”¹ 2. Attrition by Department

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department', hue='Attrition', data=df, palette='Set3')
plt.title("Attrition by Department")
plt.xticks(rotation=30)
plt.show()
```

ðŸ”¹ 3. Attrition by Overtime

```python
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='Set1')
plt.title("Attrition by Overtime Status")
plt.show()
```

ðŸ”¹ 4. Job Satisfaction Distribution

```python
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df, palette='coolwarm')
plt.title("Job Satisfaction vs. Attrition")
plt.show()
```

ðŸ”¹ 5. Monthly Income by Job Role

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='JobRole', y='MonthlyIncome', data=df)
plt.title("Monthly Income Distribution by Job Role")
plt.xticks(rotation=45)
plt.show()
```

ðŸ”¹ 6. Correlation Heatmap

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Generate heatmap
plt.figure(figsize=(14,10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='Blues', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```
